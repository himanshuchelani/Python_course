```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips=sns.load_dataset("tips")
```

```python
sns.stripplot(x=tips["tip"],color='green',jitter=False)
```

```python
sns.stripplot(x=tips["total_bill"],color='green',jitter=False)
```

```python
sns.stripplot(x='total_bill',y='day',palette="spring",data=tips,jitter=False)
```

```python
sns.stripplot(x='total_bill',y='day',palette="spring",data=tips,jitter=0.1,linewidth=1.2)
```
